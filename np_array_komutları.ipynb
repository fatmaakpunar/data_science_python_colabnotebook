```python
#matematiksel işlemler
import numpy as np
v = np.array([1,2,3,4,5])
#operatörler ile
# v/5
# v*5/10
# v**2
# v-1
#metodlar ile
np.subtract(v, 1)
np.add(v, 1)
np.mean(v)
np.sum(v)
np.min(v)
np.max(v)
np.var(v)
```
