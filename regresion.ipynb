```python
# BİR GÜNDE ORTALAMA KAÇ SAAT DERS ÇALIŞMA
saat = [[1.6], [1.65], [1.7], [1.73], [1.8]]
# SINAVDAN ALDIĞI NOTLAR
notlar = [[60], [65], [72.3], [75], [80]]
```

```python
import matplotlib.pyplot as plt
# BİR GÜNDE ORTALAMA KAÇ SAAT DERS ÇALIŞMA
saat = [[1.6], [1.65], [1.7], [1.73], [1.8]]
# SINAVDAN ALDIĞI NOT
notlar = [[60], [65], [72.3], [75], [80]]
plt.title("SAAT'E KARŞI NOTLARIN GRAFİĞİ")
plt.xlabel("SAAT")
plt.ylabel("NOTLAR")
plt.plot(saat, notlar, "k.")
# X VE Y EKSEN GENİŞLİKLERİ
plt.axis([1.5, 1.85, 50, 90])
## (1.5, 1.85, 50.0, 90.0)
plt.grid(True)
```

```python
from sklearn.linear_model import LinearRegression
# MODEL KURMA VE UYDURMA
model = LinearRegression()
model.fit(X = saat, y = notlar)
## LinearRegression()
```

```python
# TAHMİN YAPMA
notlar = model.predict([[1.75]])[0][0]
print(round(notlar, 2))
## 76.04
saat = [[1.6], [1.65], [1.7], [1.73], [1.8]]
notlar = [[60], [65], [72.3], [75], [80]]
plt.title("SAAT'E KARŞI NOTLARIN GRAFİĞİ")
plt.xlabel("SAAT")
plt.ylabel("NOTLAR")
plt.plot(saat, notlar, "k.")
plt.axis([1.5, 1.85, 50, 90])
## (1.5, 1.85, 50.0, 90.0)
plt.grid(True)
# REGRESYON DOĞRUSU ÇİZME
plt.plot(saat, model.predict(saat), color = "r")
plt.show()
```

```python
saat = [[1.6], [1.65], [1.7], [1.73], [1.8]]
notlar = [[60], [65], [72.3], [75], [80]]
plt.title("SAAT'E KARŞI NOTLARIN GRAFİĞİ")
plt.xlabel("SAAT")
plt.ylabel("NOTLAR")
plt.plot(saat, notlar, "k.")
plt.axis([1.5, 1.85, -200, 200])
## (1.5, 1.85, -200.0, 200.0)
plt.grid(True)
# REGRESYON DOĞRUSU ÇİZME
extreme_saat = [[0], [1.8]]
plt.plot(extreme_saat, model.predict(extreme_saat), color = "r")
plt.show()
```

```python
print(round(model.intercept_[0],2))
## -104.75
```

```python
print(round(model.coef_[0][0],2))
## 103.31
```

```python
import numpy as np
print("ARTIK KARELER TOPLAMI: %.2f" % np.sum((notlar - model.predict(saat)) ** 2))
## ARTIK KARELER TOPLAMI: 5.34
```

```python
# TEST VERİSİ
saat_test = [[1.58], [1.62], [1.69], [1.76], [1.82]]
notlar_test = [[58], [63], [72], [73], [85]]
```

```python
# TOPLAM KARELER TOPLAMI (TKT)
notlar_test_mean = np.mean(np.ravel(notlar_test))
TKT = np.sum((np.ravel(notlar_test) - notlar_test_mean) ** 2)
print("TKT: %.2f" % TKT)
# ARTIK KARELER TOPLAMI (HKT)
## TKT: 430.80
AKT = np.sum((np.ravel(notlar_test) - np.ravel(model.predict(saat_test))) ** 2)
print("AKT: %.2f" % AKT)
# R kare
## AKT: 24.62
R_kare = 1 - (AKT/TKT)
print("R-kare: %.2f" % R_kare)
## R-kare: 0.94
```

```python
# scikit-learn kullanarak R-kare hesaplama
print("R-kare: %.4f" % model.score(saat_test, notlar_test))
## R-kare: 0.9429
```
