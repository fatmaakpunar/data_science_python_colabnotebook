```python
import pandas as pd
import matplotlib.pyplot as plt

# Veri setini içe aktar
veri_seti = pd.read_csv('boxplotdataset.csv')

# Kutu bıyık grafiğini çizdir
plt.figure(figsize=(8, 6))  # Grafiğin boyutunu ayarla
plt.boxplot(veri_seti.values)  # Veri setinin değerlerini kullanarak kutu bıyık grafiğini çiz

# Grafiği özelleştir
plt.title('Kutu Bıyık Grafiği')  # Grafiğin başlığını ayarla
plt.xlabel('Değişkenler')  # X ekseni etiketini ayarla
plt.ylabel('Değerler')  # Y ekseni etiketini ayarla

# X ekseni üzerindeki değişken isimlerini etiketle
plt.xticks(range(1, len(veri_seti.columns) + 1), veri_seti.columns)

# Grafiği göster
plt.show()
```

```python
import matplotlib.pyplot as plt

veriler = [18.3, 18.8, 18.9, 17.2, 17.9, 16.4, 16.8, 15.3, 8.4]

plt.figure(figsize=(5, 4))
plt.boxplot(veriler, vert=False, widths=0.5, showfliers=False)

plt.title('Kutu Bıyık Grafiği')
plt.xlabel('Değerler')

# Verileri grafiğe yazdır
for i, veri in enumerate(veriler):
    plt.text(veri, 1, str(veri), ha='center', va='bottom', color='red')

plt.show()
```
