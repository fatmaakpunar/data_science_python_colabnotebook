```python
#kategorik değişken görselleştirme
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
# pd.set_option('display.widht', 500)
df = sns.load_dataset("titanic")
df.head()

df["sex"].value_counts().plot(kind='bar')
plt.show()
```

```python
#sayısal değişken görselleştirme
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
# pd.set_option('display.widht', 500)
df = sns.load_dataset("titanic")
df.head()

plt.hist(df["age"])
plt.show()

plt.boxplot(df["fare"])
plt.show()
```

```python
#matplotlib özellikleri
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)

#plot
x = np.array([1,8])
y = np.array([0,150])
#çizgi
plt.plot(x, y)
plt.show()
```

```python
#sembol
plt.plot(x,y,'*')
plt.show()
```

```python
#marker * . , x X + P s D d p H h
y = np.array([13, 28, 11, 100])
plt.plot(y, marker='*')
plt.show()
```

```python
#line kesikli
y = np.array([13, 28, 11, 100])
plt.plot(y, linestyle="dashed")
plt.show()
```

```python
#line noktalı
y = np.array([13, 28, 11, 100])
plt.plot(y, linestyle="dotted")
plt.show()
```

```python
#line nokta çizgi
y = np.array([13, 28, 11, 100])
plt.plot(y, linestyle="dashdot")
plt.show()
```

```python
#line renkli
y = np.array([13, 28, 11, 100])
plt.plot(y, linestyle="dashed", color="r")
plt.show()
```

```python
#multible lines
x = np.array([23, 18, 31, 10])
y = np.array([13, 28, 11, 100])
plt.plot(x)
plt.plot(y)
plt.show()
```

```python
#labels
x = np.array([80, 85, 90, 95, 100, 105, 110, 115,120,125])
y = np.array([240, 250,260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x, y)
#başlık
plt.title("bu ana başlık")
#x ekseni isimlendirme
plt.xlabel("x ekseni isimlendirme")
#y ekseni isimlendirme
plt.ylabel("y ekseni isimlendirme")
#ızgara
plt.grid()
```

```python
#Subplot
x = np.array([80, 85, 90, 95, 100, 105, 110, 115,120,125])
y = np.array([240, 250,260, 270, 280, 290, 300, 310, 320, 330])
plt.subplot(1,2,1)# 1 satırlık 2 sütunluk 1. grafiği
plt.title("1")
plt.plot(x, y)
```

```python
#seborn ile veri görselleştirme seaborn daha az çabayla veri görselleştirir
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset("tips")
df.head()

df["sex"].value_counts()
sns.countplot(x=df["sex"], data=df)
plt.show()

#matplotlibde nasıl yapıyorduk
df["sex"].value_counts().plot(kind='bar')
plt.show()
```

```python
#seabornla sayısal değişkenler
sns.boxplot(x=df["total_bill"])
plt.show()
```

```python
#histogram
df["total_bill"].hist()
plt.show()
```
