```python
#koşullu seçim işlemleri
import pandas as pd
import seaborn as sns
pd.set_option('display.max_columns', None)
df = sns.load_dataset("titanic")
df.head()

df[df["age"]>50].head()#yaşı elliden büyük olan ilk 5 kişi
df[df["age"]>50]["age"].count()#yaşı elliden büyük kaç kişi var,64
df.loc[df["age"]>50, "class"].head()#yaşı elliden büyük olanların sınıf bilgisi
df.loc[df["age"]>50, ["age","class"]].head()#yaşı elliden büyük olanların hem yaşı hem sınıf bilgisi
df.loc[(df["age"]>50) & (df["sex"] == "male"), ["age","class"]].head()#yaşı elliden büyük ve erkek olanların yaş ve sınıf bilgisi
```

```python
#toplulaştırma ve gruplama
#toplulaştırma fonksiyonları:
#count()
# first()
# last()
# mean()
# median()
# min()
# max()
# std()
# var()
#sum()
#pivot-table
import pandas as pd
import seaborn as sns
pd.set_option('display.max_columns', None)
df = sns.load_dataset("titanic")
df.head()

df["age"].mean()#yaş ortalaması
df.groupby("sex")["age"].mean()#cinsiyete göre yaş değişkeninin ortalaması
df.groupby("sex").agg({"age":"mean"})#aynı işlevi yapar ama üstünde daha fazla işlem uygulayabiliriz 
df.groupby("sex").agg({"age":["mean", "sum"],
                       "survived": "mean"})#yaklaşık olarak kadınların yüzde 74 ü erkeklerin yüzde 18 i hayatta kalmış

```

```python
df.groupby(["sex", "embark_town","class"]).agg({"age":["mean", "sum"],
                       "survived": "mean"})#cinsiyete ve bniş limanına ve sınıfa  göre hayatta kalma olsılığı
```

```python
df.groupby(["sex", "embark_town","class"]).agg({"age":["mean", "sum"],
                       "survived": "mean",
                       "sex":"count"})#bu olasılıktakiler kaç kişi countla bakarız
```

```python
#pivot table
import pandas as pd
import seaborn as sns
pd.set_option('display.max_columns', None)
df = sns.load_dataset("titanic")
df.head()

df.pivot_table("survived", "sex", "embarked")#bize tablo oluşturur pivot tablo
```

```python
#yaş değişkeni hayatta kalmayla nasıl ilşkilendirilir
df["new_age"] = pd.cut(df["age"],[0,10,18,25,40,90])
df.pivot_table("survived","sex",["new_age","class"])#yaşayanların yaş ve sınıflara göre gruplanmış cinsiyetlere göre olasılığı

```

```python
#Apply ve Lambda
import pandas as pd
import seaborn as sns
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)
df = sns.load_dataset("titanic")
df.head()

df["age2"] = df["age"]*2
df["age3"] = df["age"]*5
df[["age","age2","age3"]].apply(lambda x: x/10).head()#yaş değişkenlerini 10 a böl
#apply satır sütun işlemleri ve fonksiyonları daha kolay yapmamızı sağlar
```

```python
#birleştirme (join) işlemleri
import numpy as np
import pandas as pd
m = np.random.randint(1 , 30 , size=(5,3))
df1 =  pd.DataFrame(m, columns=["var1", "var2", "var3"])#dataframe oluşturduk
df2 = df1 + 99
pd.concat([df1, df2], ignore_index=True)
```

```python
#merge ile birleştirme işlemi
pd.merge(df1,df2)#şeklinde yapılır
```
